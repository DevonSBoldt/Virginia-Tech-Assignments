**Find Extremum: Identify the most populated countries in 2020 and visualize them in a bar chart.**

### **Data Dimensions:**  
- **Country** (Categorical, X-axis)  
- **Population 2020** (Quantitative, Y-axis)  

### **Mapping to Visualization Components:**  
- **X-axis:** Country names (categorical)  
- **Y-axis:** Population size (bar height represents population)  
- **Color:** Bars are colored using a gradient palette (e.g., `viridis`) to enhance readability  

### **Why this mapping?**  
- The bar height effectively communicates population size differences.  
- Sorting countries by population ensures an intuitive comparison.  

**How does this visualization support this task?**<br><br>
This bar chart allows users to quickly identify the most populated countries and compare their relative sizes. The height of each bar visually represents the population count, making it easy to see how China and India dominate in population compared to other countries.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
file_path = "population_100.csv"
df = pd.read_csv(file_path)

# Convert population column to numeric if not already
df['Population 2020'] = pd.to_numeric(df['Population 2020'], errors='coerce')

# Select the top 10 most populated countries
top_10_population = df.nlargest(10, 'Population 2020')

# Create bar chart
plt.figure(figsize=(12, 6))
sns.barplot(data=top_10_population, x='Country', y='Population 2020', palette='viridis')
plt.xticks(rotation=45)
plt.ylabel("Population (2020)")
plt.title("Top 10 Most Populated Countries")
plt.show()
```

# Analyze Population Density (Scatter Plot)

### **Data Dimensions:**  
- **Country** (Categorical, X-axis)  
- **Density People per km²** (Quantitative, Y-axis)  
- **Region** (Categorical, Hue/Color)  

### **Mapping to Visualization Components:**  
- **X-axis:** Country names (categorical)  
- **Y-axis:** Population density (higher positions indicate higher density)  
- **Color:** Different regions are assigned unique colors for categorization  
- **Alpha (Transparency):** Adjusted to prevent visual clutter with overlapping points  

### **Why this mapping?**  
- Helps identify which countries have the highest population density.  
- Regional coloring allows users to observe density trends by continent.  


**How does this visualization support this task?**<br><br>

This scatter plot maps each country's population density (people per km²), categorized by region. It highlights highly crowded areas, such as South and East Asia, while also showing regions with lower densities like North America. Users can easily identify trends and outliers in population distribution.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Convert density column to numeric if not already
df['Density People per km^2'] = pd.to_numeric(df['Density People per km^2'], errors='coerce')

# Create scatter plot
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df, x='Country', y='Density People per km^2', hue='Region', alpha=0.7)
plt.xticks(rotation=90, fontsize=8)
plt.ylabel("Population Density (People per km²)")
plt.title("Population Density by Country and Region")
plt.legend(title="Region", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

**Correlate: Examine the relationship between yearly population growth rates and the top 10 most populated countries through a line chart.**

# Track Yearly Population Change (Line Chart)

### **Data Dimensions:**  
- **Country** (Categorical, X-axis)  
- **Yearly Change (%)** (Quantitative, Y-axis)  

### **Mapping to Visualization Components:**  
- **X-axis:** Country names (categorical)  
- **Y-axis:** Yearly population change percentage (higher values indicate faster growth)  
- **Marker Style:** Circles (`o`) added at each data point for clarity  
- **Color:** Single color (`blue`) to focus on trend patterns  

### **Why this mapping?**  
- A line chart shows trends across countries, making it easier to compare growth rates.  
- Data points emphasize actual recorded values rather than just trend direction.  


**How does this visualization support this task?**<br><br>

The line chart visually tracks the yearly population growth rate for the top 10 most populated countries. By connecting data points, users can identify increasing or decreasing trends over time. Countries with higher birth rates and migration gains stand out, while those with slower growth can be examined for demographic changes.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Convert Yearly Change to numeric
df['Yearly Change'] = df['Yearly Change'].str.rstrip('%').astype(float)

# Select the top 10 most populated countries
top_10_population = df.nlargest(10, 'Population 2020')

# Create line chart
plt.figure(figsize=(12, 6))
sns.lineplot(data=top_10_population, x='Country', y='Yearly Change', marker="o", color='b')
plt.xticks(rotation=45)
plt.ylabel("Yearly Population Change (%)")
plt.title("Yearly Population Growth Rate for Top 10 Countries")
plt.show()
```

**Correlate: Compare land area and total population to explore their relationship, using a scatter plot with a logarithmic scale for improved visibility.**

# Compare Land Area vs. Population (Scatter Plot)

### **Data Dimensions:**  
- **Land Area km²** (Quantitative, X-axis)  
- **Population 2020** (Quantitative, Y-axis)  
- **Region** (Categorical, Hue/Color)  

### **Mapping to Visualization Components:**  
- **X-axis:** Land area in km² (log scale to accommodate large variations)  
- **Y-axis:** Total population in 2020 (log scale to emphasize differences)  
- **Color:** Different colors based on **Region** to indicate continental groupings  
- **Alpha (Transparency):** Adjusted to reduce overlapping effects  

### **Why this mapping?**  
- A scatter plot helps analyze whether large land area corresponds to higher population.  
- The log scale ensures that smaller values remain visible alongside larger ones.  
- Regional color coding provides geographic context to the analysis.  


**How does this visualization support this task?**<br><br>

This scatter plot maps total land area (X-axis) against total population (Y-axis) with a logarithmic scale. It helps users see whether large countries necessarily have large populations and vice versa. The visualization clarifies that some small countries (like Bangladesh) have high population densities, while large countries (like Russia) have relatively low population counts.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Convert necessary columns to numeric
df['Land Area km^2'] = pd.to_numeric(df['Land Area km^2'], errors='coerce')

# Create scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Land Area km^2', y='Population 2020', hue='Region', alpha=0.7)
plt.xscale('log')  # Log scale for better visualization
plt.yscale('log')  # Log scale for better visualization
plt.xlabel("Land Area (km²)")
plt.ylabel("Total Population (2020)")
plt.title("Land Area vs. Total Population")
plt.legend(title="Region", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
