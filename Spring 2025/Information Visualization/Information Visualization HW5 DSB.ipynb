**Visualization For Good**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = "broadway.csv"
df = pd.read_csv(file_path)

# Aggregate total attendance by show type
df_attendance = df.groupby("Show.Type")["Statistics.Attendance"].sum()

# Define the color palettes
color_palettes = [
    ["#6A5ACD", "#4682B4", "#48D1CC"],  # Cool and Calming
]

# Generate and display the pie charts
for i, colors in enumerate(color_palettes):
    fig, ax = plt.subplots(figsize=(8, 8))
    wedges, _, autotexts = ax.pie(
        df_attendance, labels=None, autopct="%1.1f%%",
        startangle=90, wedgeprops={"edgecolor": "black"},
        textprops={"fontsize": 14}, colors=colors  # Apply custom colors
    )

    # Make the percentage labels bold for better visibility
    for autotext in autotexts:
        autotext.set_fontsize(14)
        autotext.set_fontweight("bold")

    # Add a legend
    ax.legend(wedges, df_attendance.index, title="Show Type", loc="upper right", fontsize=12)

    # Formatting
    ax.set_title(f"Broadway Attendance Distribution", fontsize=16)

    # Show the plot
    plt.show()
```

**Visualization for Evil**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = "labor.csv"
df = pd.read_csv(file_path)

# Intentionally misleading data transformation: Squaring the unemployment rate to exaggerate differences
df['Manipulated Unemployment Rate'] = df['Data.Unemployed.Black or African American.Unemployment Rate.All'] ** 2

# Creating an 'evil' visualization
plt.figure(figsize=(12, 6))

# Intentionally misleading plot
plt.plot(df['Time.Year'], df['Manipulated Unemployment Rate'], label='Black Unemployment', linestyle='--', color='red', linewidth=3)
plt.plot(df['Time.Year'], df['Data.Unemployed.White.Unemployment Rate.All'], label='White Unemployment', linestyle='-', color='blue', linewidth=3)

# Misleading Y-axis scale
plt.ylim(0, max(df['Manipulated Unemployment Rate']) * 1.2)

# Skewed title to influence perception
plt.title("Who Gets Left Behind? The Alarming Truth About Unemployment in America!", fontsize=14, fontweight='bold')

# Misleading labels
plt.xlabel("Time (Years)", fontsize=12, fontweight='bold')
plt.ylabel("Unemployment Rate (Manipulated)", fontsize=12, fontweight='bold')

# Hiding grid to make interpretation harder
plt.grid(False)

# Overcrowded legend placement
plt.legend(loc='upper left', fontsize=10)

# No data source provided
plt.show()
```
